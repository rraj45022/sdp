```python
y=df.Result
X=df.drop('Result', axis=1)
```

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import joblib
import time

# Step 1: Split Data
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

start_time = time.time()

# Step 2: Upsample Training Data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Step 3: Hyperparameter Tuning with Gradient Boosting and k-fold Cross Validation
param_grid = {
    'n_estimators': [6000],
    'learning_rate': [0.01],
    'max_depth': [8],
    'subsample': [0.8],
    'verbose': [1],
    
}

gb_classifier = GradientBoostingClassifier(random_state=42)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
grid_search = GridSearchCV(gb_classifier, param_grid, cv=kf, scoring='accuracy')
grid_search.fit(X_train_resampled, y_train_resampled)

# Best hyperparameters
best_params = grid_search.best_params_

# Step 4: Model Evaluation
best_gb_classifier = grid_search.best_estimator_
y_pred_test = best_gb_classifier.predict(X_test)
accuracy_test = accuracy_score(y_test, y_pred_test)

# Calculate precision
precision = precision_score(y_test, y_pred_test)

# Calculate recall
recall = recall_score(y_test, y_pred_test)

# Calculate F1-score
f1 = f1_score(y_test, y_pred_test)

# Calculate confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()

# Calculate specificity
specificity = tn / (tn + fp)

# Calculate false negative rate
fnr = fn / (fn + tp)

# Calculate false discovery rate
fdr = fp / (fp + tp)

# Calculate misclassification rate
misclassification_rate = (fp + fn) / (tp + tn + fp + fn)

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_test)
roc_auc = auc(fpr, tpr)

end_time = time.time()
total_runtime_gdb_wfe = end_time - start_time
print("Total Runtime :", total_runtime_gdb_wfe, "seconds")


print("Best Hyperparameters:", best_params)
print("Test Accuracy:", accuracy_test)


# Print the metrics
print("Accuracy:", accuracy_test)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Specificity:", specificity)
print("False Negative Rate:", fnr)
print("False Discovery Rate:", fdr)
print("Misclassification Rate:", misclassification_rate)
print("AUC:", roc_auc)


# Step 1: Save the best model
model_filename = 'gradient_boosting_model_wfe.pkl'
joblib.dump(best_gb_classifier, model_filename)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, roc_curve, auc
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
from scikeras.wrappers import KerasClassifier
import joblib

# Step 2: Upsample Training Data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_selected, y_train)


# Step 4: Hyperparameter Tuning with MLP and k-fold Cross Validation
param_dist_mlp = {
    'batch_size': [32, 64],
    'epochs': [40],
    'optimizer': ['adam'],
}

# Define MLP model function with flexible hyperparameters
def create_mlp_model(optimizer='adam', activation='relu', units=64):
    model = Sequential()
    model.add(Dense(units=units, activation=activation, input_dim=X_train_resampled.shape[1]))
    model.add(Dense(units=units // 2, activation=activation))
    model.add(Dense(units=units // 4, activation=activation))  # Additional layer 1
    model.add(Dense(units=units // 8, activation=activation))  # Additional layer 2
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

mlp_classifier = KerasClassifier(build_fn=create_mlp_model, verbose=0, units=256, activation='relu')
kf = KFold(n_splits=10, shuffle=True, random_state=42)
random_search_mlp = RandomizedSearchCV(estimator=mlp_classifier, 
                                        param_distributions=param_dist_mlp, 
                                         # Number of random parameter combinations to try
                                        cv=kf, 
                                        scoring='accuracy',
                                        random_state=42)
random_search_mlp.fit(X_train_resampled, y_train_resampled)

# Best hyperparameters
best_params_mlp = random_search_mlp.best_params_

# Step 5: Model Evaluation
best_mlp_classifier = random_search_mlp.best_estimator_
y_pred_test_mlp = best_mlp_classifier.predict(X_test_selected)
accuracy_test_mlp = accuracy_score(y_test, y_pred_test_mlp)

precision = precision_score(y_test, y_pred_test_mlp)

# Calculate recall
recall = recall_score(y_test, y_pred_test_mlp)

# Calculate F1-score
f1 = f1_score(y_test, y_pred_test_mlp)

# Calculate confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test_mlp).ravel()

# Calculate specificity
specificity = tn / (tn + fp)

# Calculate false negative rate
fnr = fn / (fn + tp)

# Calculate false discovery rate
fdr = fp / (fp + tp)

# Calculate misclassification rate
misclassification_rate = (fp + fn) / (tp + tn + fp + fn)

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_test_mlp
                                )
roc_auc = auc(fpr, tpr)


print("Best Hyperparameters (MLP):", best_params_mlp)
print("Test Accuracy (MLP):", accuracy_test_mlp)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Specificity:", specificity)
print("False Negative Rate:", fnr)
print("False Discovery Rate:", fdr)
print("Misclassification Rate:", misclassification_rate)
print("AUC:", roc_auc)

# Step 1: Save the best model
model_filename = 'best_mlp_classifier_model1.pkl'
joblib.dump(best_mlp_classifier, model_filename)
```
